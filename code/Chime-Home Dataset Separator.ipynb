```python
import pandas as pd
import os
import shutil
```

```python
for filename in os.listdir("./chunks/"):
    if filename.endswith(".16kHz.wav"):
        file = filename[:-9]
        info = pd.read_csv("./chunks/" + file + "csv")
        label = str(info.iloc[8][1])

        if "nan" in label:
            continue

        original = r'./chunks/' + filename
        
        if "m" in label:
            # save audio to "./Separated/Male_Speech/"
            target = r'./Separated/Male_Speech/' + filename
            shutil.copyfile(original, target)
        elif "f" in label:
            # save audio to "./Separated/Female_Speech/"
            target = r'./Separated/Female_Speech/' + filename
            shutil.copyfile(original, target)
        elif "c" in label:
            # save audio to "./Separated/Child_Speech/"
            target = r'./Separated/Child_Speech/' + filename
            shutil.copyfile(original, target)
        else:
            # save audio to "./Separated/Other/"
            target = r'./Separated/Other/' + filename
            shutil.copyfile(original, target)        
```
